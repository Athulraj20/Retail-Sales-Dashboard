```python
# Retail Sales Analysis

## Objective

Analyze retail sales data to identify top-performing products, customer segments, and revenue-driving categories.

Tools Used:
- Python
- Pandas
- Power BI
```

```python
import pandas as pd

df = pd.read_csv("Retail_Analysis_Final.csv")

df.head()
```

```python
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

df.info()
```

```python
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

df.info()
```

```python
print("Total Revenue:", round(df['Product_Sales'].sum(),2))
print("Total Orders:", df['Order_ID'].nunique())
print("Total Products:", df['Product_Standard'].nunique())
```

```python
df.groupby('Product_Category')['Product_Sales']\
.sum()\
.sort_values(ascending=False)\
.head(10)
```

```python
df.groupby('Product_Standard')['Product_Sales']\
.sum()\
.sort_values(ascending=False)\
.head(10)
```

```python
df.groupby('Customer_Type')['Product_Sales']\
.sum()\
.sort_values(ascending=False)
```

```python
## Key Insights

- Rice generated the highest category revenue.
- Banana products showed strong demand.
- Retail Price customers contributed the largest revenue share.
- Coconut and Edible Oil categories were strong performers.
```
